```python
import networkx as nx
opts = {"with_labels": True, "node_color":'g'}
```

```python
n = 10
G = nx.path_graph(n) #Crea un path (camino)
nx.draw(G, **{"with_labels": True,"node_color":'y'})# Path creado originalmente
for v in range(n-2):
    G.add_edge(v,v+2)# crea una conexion entre el nodo v y su vecino v+2,pq el v+1 ya lo tiene conectado
nx.draw_circular(G,**opts)#Path despues de agregar conexion entre un nodo y su vecino n+2
# Aqui no se crea conexion entre el ultimo nodo y el primero
```

```python
n = 10
G = nx.cycle_graph(n)# grafo circular
#nx.draw(G,**opts)
for v in G:
    G.add_edge(v,(v+2) % n)# crea una conexion entre el nodo v y su vecino v+2,y  conecta el ultimo nodo con el primero
    print((v+2) % n)
nx.draw_circular(G,**opts)
```

```python
#Esta funcion crea una red conectando cada nodo a sus 2k vecinos
def create_network(n, k):
    G = nx.cycle_graph(n)
    for v in G:
        for o in range(2, k+1): #decide hasta que nodo se va a conectar
            G.add_edge(v, (v+o) % n)#crea la conexion
    return G
G = create_network(10,2)
nx.draw(G,**opts)
```
